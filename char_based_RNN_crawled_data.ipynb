```
import tensorflow as tf

import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from sklearn.model_selection import train_test_split

import unicodedata
import re
import numpy as np
import os
import io
import time
```

```
def extract_lines(txt_name):
  sentences = []
  with open(("/content/drive/MyDrive/applied_ai_hw3/" + str(txt_name)), "r") as f:
    for line in f:
        for word in line.strip().split("."):
            if len(word) > 15:
              sentences.append(word)
  return sentences
```

```
# Converts the unicode file to ascii
def unicode_to_ascii(s):
    mylist = []
    for c in unicodedata.normalize('NFD', s):
        if unicodedata.category(c) != 'Mn':
            mylist.append(c)
    
    return ''.join(mylist)
```

```
#TODO: arguement name should be "sentence"
def preprocess_sentence(w):
    w = unicode_to_ascii(w.lower().strip())

    # creating a space between a word and the punctuation following it
    # eg: "he is a boy." => "he is a boy ."
    # Reference:- https://stackoverflow.com/questions/3645931/python-padding-punctuation-with-white-spaces-keeping-punctuation
    w = re.sub(r"([?.!,¿])", r" \1 ", w)
    w = re.sub(r'[" "]+', " ", w)

    # replacing everything with space except (a-z, A-Z, ".", "?", "!", ",")
    w = re.sub(r"[İı]", "i", w)
    w = re.sub(r"[\n]", "", w)
    w = re.sub(r"[^a-zA-Z?.!,¿]+", " ", w)

    w = w.strip()

    # adding a start and an end token to the sentence
    # so that the model know when to start and stop predicting.

    return w
```

```
lines = []
```

```
def add_lines(txt_name):
  sentences = extract_lines(txt_name)
  for sentence in sentences:
      lines.append(preprocess_sentence(sentence))
```

```
add_lines("ensonhaber.txt")
print(len(lines))
add_lines("ahaber.txt")
print(len(lines))
add_lines("ntv.txt")
print(len(lines))
add_lines("hurriyet.txt")
print(len(lines))
```
